# Credit Card Fraud Data Cleaning 

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Hp\Documents\jobs\Annie Jobs\Elizabeth Job\Fraud Detection\archive (1)\fraudTest.csv")
print(df.head())
```

Checking for duplicates

```python
df.drop_duplicates()
```

Dropping the unnamed column 

```python
# Dropping the unnamed column

if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'])
    df = df.drop(columns=['Unnamed: 0'])
    
```

```python
#Check for duplicate rows

duplicates = df.duplicated().sum()
print(df.duplicated().sum())
```

```python
#Check for missing values

print("Missing Values:")
print(df.isnull().sum())
```

```python
#Check for unique values in gender

print("Unique Values in 'gender':")
print(df['gender'].unique())
```

```python
#Check for unique values 'is_fraud column'

print("Unique Values in 'is_fraud':")
print(df['is_fraud'].unique())
```

```python
# Standardize the date format

df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
df['dob'] = pd.to_datetime(df['dob'], format='%Y-%m-%d', errors='coerce')

print(df['trans_date_trans_time'].head())
```

```python
# Separating the date and time into new columns

df['trans_date'] = df['trans_date_trans_time'].dt.date
df['trans_time'] = df['trans_date_trans_time'].dt.time
```

```python
# Create new columns for date and time

df['date_of_birth'] = df['dob'].dt.date
```

```python
# Drop the original 'trans_date_trans_time' column

df.drop(columns=['trans_date_trans_time'], inplace=True)
```

```python
# Drop the original 'dob' column

df.drop(columns=['dob'], inplace=True)

print(df.head())
```

```python
# Rearrange the column order

new_column_order = [
    'trans_date',
    'trans_time',
    'cc_num',
    'merchant',
    'category',
    'amt',
    'first',        
    'last',         
    'date_of_birth',
    'gender',
    'street',
    'city',
    'state',
    'city_pop',
    'job',
    'trans_num',
    'unix_time',
    'merch_lat',
    'merch_long',
    'is_fraud'
]

# Reorder the DataFrame columns
df = df[new_column_order]
```
